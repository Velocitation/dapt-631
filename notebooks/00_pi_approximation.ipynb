**Numerical approximation of π:** 

As points are randomly scattered inside the unit square, some fall within the unit circle. The fraction of points inside the circle approaches π/4 as points are added.
[From Wikipedia](https://en.wikipedia.org/wiki/Approximations_of_%CF%80#/media/File:Pi_30K.gif).

Code Source: [@__mharrison__](https://twitter.com/__mharrison__/status/1503402735361699840)

```python
import random
import matplotlib.pyplot as plt
```

```python
est = []
in_circle = 0
xs, ys, ins = [], [], []
random.seed(314)
```

```python
for i in range(50000):
    # generate a random point in 2D space
    x = random.random()
    y = random.random()
    # calculate its distance from the origin
    radius = (x**2 + y**2)**.5
    
    if radius <= 1: 
        # the random point is within the circular arc
        in_circle += 1
        ins.append(1)
    else:
        # the random point is outside the circular arc
        ins.append(0)
    
    xs.append(x)
    ys.append(y)
    
    # ratio of points that fell within the circular arc
    ratio = 4 * in_circle / len(xs)
    est.append(ratio)
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(10, 5))

axes[0].scatter(xs, ys, alpha=.01, c=ins)
axes[1].plot(est)
axes[1].axhline(3.14, lw=.5, c='tomato')
axes[1].set_ylim(3.08, 3.18);
```

```python
print(f'The approximate value of \u03C0 is: {est[-1]}')
```
